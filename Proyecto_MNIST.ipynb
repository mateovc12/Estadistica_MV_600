#Reconocimiento de Imágenes con MNIST


La base de datos MNIST es una colección de 70,000 imágenes de dígitos escritos a mano, dividida en 60,000 imágenes para entrenamiento y 10,000 para pruebas. Cada imagen tiene un tamaño de 28x28 píxeles en escala de grises.

Esta puede usarse para entrenar Modelos de Machine Learning que sean capaces de reconocer y clasificar dígitos con características especiales.

Las muestras incluidas en el conjunto de entrenamiento fueron el resultado de escanear dígitos manuscritos de 250 personas (estudiantes de preparatoria y empleados de la oficina del Censo de los Estados Unidos). El dataset de pruebas contiene dígitos escaneados de otras 250 personas diferentes (con los mismos perfiles), lo que permite asegurar que los modelos obtenidos son capaces de interpretar dígitos incluso de personas no involucradas en la generación de los datos de entrenamiento.

##Visualización de datos de MNIST
La base de entrenamiento tiene la siguiente estructura en forma de tabla:

*Nota* : La base de datos de entrenamiento en csv se descarga en el siguiente link https://www.interactivechaos.com/data/mnist.zip

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("mnist.csv", header = None)
data.head()
```

En la columna de índice 0  se muestra la etiqueta del dígito que contiene cada muestra. Estos son dígitos del 0 al 9.

Después la columna 1 contiene el píxel de la esquina superior izquierda de cada imagen. La columna de índice 28 contiene el píxel de la esquina superior derecha, y la columna de índice 29 contiene el píxel izquierdo de la segunda fila de la imagen. Y así sucesivamente. La imagen se completa mediante la activación de pixeles por filas y columnas.

A continuación se muestra un ejemplo de la forma en que se buscan y visualizan los distintos elementos de MNIST:

```python
#Código para visualizar algunos de los datos a la vez

from tensorflow.keras.datasets import mnist

# Cargar el dataset MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Visualizar algunas imágenes del dataset
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f'Label: {y_train[i]}')
    plt.axis('off')
plt.show()
```

Estas técnicas se utilizan para extraer información útil de las imágenes y tomar decisiones basadas en esa información. Y se utiliza en una amplia variedad de aplicaciones y campos.

2. Industria y manufactura:

4. Ciencias ambientales:

**Teledetección:** Análisis de imágenes satelitales para estudiar el cambio climático, el uso del suelo y los desastres naturales.

5. Entretenimiento:

**Edición de imágenes:** Mejoras y modificaciones de fotografías y videos.

6. Reconocimiento de patrones y análisis de datos:
